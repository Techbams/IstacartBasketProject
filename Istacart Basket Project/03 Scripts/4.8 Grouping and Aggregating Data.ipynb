## 05. Aggregating Data with transfer()

```python
# ords_prods_merge
```

Layalty flag
- if max orders the user has made is over 40, then the customer will be labelled a "Loyal Customer"
- if max orders the user has made is over 10 but less than or equal to 40, then the customer will be labelled a "Regular Customer"
- if max orders the user has made is less than 10, then the customer will be labelled a "New Customer"

1. split data into groups based off user_id
2. Apply the transform() on order_number column - max orders for each user
3. Create a new column, "max_order", 

```python
ords_prods_merge.head(2)
```

```python
ords_prods_merge['max_order'] =  ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## 06. Deriving Columns with Loc ()

Layalty flag
- if max orders the user has made is over 40, then the customer will be labelled a "Loyal Customer"
- if max orders the user has made is over 10 but less than or equal to 40, then the customer will be labelled a "Regular Customer"
- if max orders the user has made is less than or equal 10, then the customer will be labelled a "New Customer"

```python
ords_prods_merge.loc[ ords_prods_merge['max_order'] > 40 , 'loyalty_flag' ] = 'Loyal customer'
```

```python
ords_prods_merge.loc[ (ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10) , 'loyalty_flag' ] = 'Regular customer'
```

```python
ords_prods_merge.loc[ ords_prods_merge['max_order'] <= 10, 'loyalty_flag'   ] = 'New customer'
```

```python
# check frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
