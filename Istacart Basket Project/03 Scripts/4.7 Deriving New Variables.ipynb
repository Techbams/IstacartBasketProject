
- if item's price is lower than or equal to $5, it will be labelled a "low_range product."
- if item's price is above $5 but lower than or equal to $15, it will be labelled a "mid-range product."
- if item's price is greater than $15, it will be labelled a "high_range product."
 

```python
ords_prods_merge = ords_prods_merge.drop(columns = '_merge')
```

```python
# generate first 1000000 rows of data from the ords_prods_merge dataframe to test the function
df = ords_prods_merge[:1000000]
```

```python
df.head()
```

```python
# define a function to capture the price label requirement
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

```python
# apply the function to a dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

# Task

```python
## Assessing the orders column

ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
## Creating periods of time 
## Hours 10,11,12,13,14,15,16 will be called as Most orders
## Hours 0,1,2,3,4,5,6 will be called as Fewest orders
## Rest of the hours will be called as Average orders

hours = []

for value in ords_prods_merge['order_hour_of_day']:
    if (value >=10) and (value <= 16):
        hours.append('Most orders')
    elif (value >=0) and (value <= 6):
        hours.append('Fewest orders')
    else: 
        hours.append('Average orders')
```

```python
ords_prods_merge['busiest_period_of_day'] = hours

ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

# Task - busiest period of the day column (Alternative approach)

```python
# Create a for-loop if statement using the above breakdown
hour_of_day = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
# Add result to dataframe as "busiest_period_of_day" column
ords_prods_merged['busiest_period_of_day'] = hour_of_day
```
